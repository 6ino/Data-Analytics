Is KMeans a supervised or an unsupervised method?
As you will notice the cluster generated by KMeans is not a good one.
Find a better way to cluster the same dataset and plot the clusters.

```python
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN

X,y = make_circles(n_samples=1500, factor=.3,noise=.05) 
plt.scatter(X[:,0], X[:,1]); 
dbscan = DBSCAN(eps=0.3, min_samples=5) 
labels = dbscan.fit_predict(X) 
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis') 
plt.show()

#KMeans is an unsupervised method. This method above is a better way to cluster.
```

**Question 8**. The transformed image above is such that the flower object is almost separate from the background. What is this type of image processing task called? Use another image with multiple objects on it and perform similar transformation using KMeans where objects will be separated from the background.

```python
from sklearn.datasets import load_sample_image
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Load a cool city pic
image = load_sample_image("china.jpg")  #

#Reshaping
X = (image / 255.0).reshape(-1, 3)  # Don't forget to normalize

# number of colorts
n_colors = 5

# top colors
kmeans = KMeans(n_clusters=n_colors)
kmeans.fit(X)
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Rebuild the image with the main colors
segmented_img = cluster_centers[labels].reshape(image.shape)
segmented_img = (255 * segmented_img).astype(np.uint8)  # Scale back up

# Let's show the before and after
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('OG Image')

plt.subplot(1, 2, 2)
plt.imshow(segmented_img)
plt.title(f'Segmented with {n_colors} colors')

plt.show()
```
